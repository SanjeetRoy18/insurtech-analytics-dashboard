```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
# 1. Load the dataset 
df = pd.read_csv(r"C:\Users\sanje\MISC_\Downloads\MAHE\Study_Material\Xceedance\data\Employee_Insurance_and_Benefits_Dataset.csv")
print(df.head())
```

```python
# Set up the plotting style for clean corporate visuals
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# CHART 1: BAR GRAPH - Average Claim Amount by Department
plt.figure()
sns.barplot(x='Department', y='Total Claim Amount ($)', data=df, estimator='mean', errorbar=None, palette='Blues_d')
plt.title('Average Claim Amount by Department', fontsize=14, fontweight='bold')
plt.ylabel('Average Claim Amount ($)')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
#CHART 2: PIE CHART - Distribution of Insurance Plan Tiers
plt.figure()
tier_counts = df['Insurance Plan Tier'].value_counts()
plt.pie(tier_counts, labels=tier_counts.index, autopct='%1.1f%%', startangle=180, colors=sns.color_palette('pastel'))
plt.title('Distribution of Insurance Plan Tiers', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#CHART 3: SCATTER PLOT - Annual Premium vs. Total Claim Amount
plt.figure()
sns.scatterplot(data=df, x='Annual Premium ($)', y='Total Claim Amount ($)', hue='Office Location', style='Office Location', s=100)
plt.title('Annual Premium vs. Total Claim Amount', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
## CHART 4: HEATMAP - Insurance Metrics Correlation Matrix
plt.figure(figsize=(8, 6))
# Select only numeric columns for correlation matrix
numeric_cols = ['Annual Premium ($)', 'Wellness Stipend Used ($)', 'Claim Count (YTD)', 'Total Claim Amount ($)']
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Insurance Metrics Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# CHART 5: HISTOGRAM - Distribution of Wellness Stipend Utilization
plt.figure()
sns.histplot(data=df, x='Wellness Stipend Used ($)', bins=10, kde=True, color='orange')
plt.title('Distribution of Wellness Stipend Utilization', fontsize=14, fontweight='bold')
plt.xlabel('Stipend Amount Used ($)')
plt.ylabel('Employee Count')
plt.tight_layout()
plt.show()
```
